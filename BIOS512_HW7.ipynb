## Question 2
#### Explain how patches relate to the idea of committing a change to the history of your code?

Git uses the patch program which makes patch files (temporary documents) containing different versions of code. These changes can be committed to the history of code and the patch files will store the changes. 

## Question 3
#### What is the difference between a branch and a fork?

A branch is the history of your code up until a certain moment (also called the head or the history). It contains the history of patch files either going back to the beginning of the repository or back to when the branch diverged.

A fork different from a branch because it is a permanently separate copy of the original repository. Branches are  temporary and are part of the same repository. 

## Question 4
#### How are diff and patch related? What’s the difference?

diff shows the difference between two files of code in both a human-readable and machine-readable way. This is related to patch because the Patch program can take the diff and help transform one file into the second file. Both diff and patch are being used by git to model the source code. However, they are different because diff is reporting the difference between files while patch stores the actual versions of code and can use the diff to transform files from and old one to a more updated one.
